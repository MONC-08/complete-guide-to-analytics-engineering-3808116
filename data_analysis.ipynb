```python
import pandas as pd
df = pd.read_csv('red_30_tech_us_sales.csv')
```

```python
df.columns = df.columns.str.strip()
grouped_df = df.groupby('EmpID')['Quantity'].sum()
sorted_df = grouped_df.sort_values(ascending = False)
sorted_df.head(10)
```

```python
# alternate way of sorting

top_employees = grouped_df.nlargest(10)   # decending order 
top_employees
```
