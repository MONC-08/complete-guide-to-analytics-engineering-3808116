```python
import pandas as pd

df = pd.read_csv('red_30_tech_us_sales.csv')
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
df.head()
```

```python
df['Order Total'] = df['Order Total'].astype(str).str.replace(r'[\s"$#,]', '', regex = True).astype(float)
df['Discount'] = df['Discount'].astype(str).str.replace(r'[\s"$#,]', '', regex = True).astype(float)
df['Price'] = df['Price'].astype(str).str.replace(r'[\s"$#,]', '', regex = True).astype(float)
```

```python
df.head()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'],
                  format = '%m/%d/%Y',
                  errors = 'coerce'
)
```

```python
# setting a flag
df['OrderDate_null'] = df['OrderDate'].isnull()
df.head()
```

```python
conditon = df['Quantity'].isnull() & (df['OrderDate_null'] == False)
df.loc[conditon, 'Quantity'] = (df['Order Total'] + df['Discount']) / df['Price']
```
